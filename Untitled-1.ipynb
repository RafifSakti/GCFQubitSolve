```python
def test2(ng):
    N=300
    Ej=0.2
    alpha=1

    Ec=Ej*alpha
    # ng=0
    S=20
    S=np.pi

    H=np.zeros([N,N],dtype=complex)
    delta=np.linspace(-S,S,num=N)
    d=delta[1]-delta[0]

    D=np.zeros([N,N],dtype=complex)
    for i in range(N):
        D[i,i]= 8*Ec/d**2 +4*Ec*ng**2 -Ej*np.cos(delta[i])#1 +delta[i]**2/2 -delta[i]**4/24 #+delta[i]**6/720
        # D[i,i]= 8*alpha/d**2 +4*alpha*ng**2  -1 +delta[i]**2/2 -delta[i]**4/24 #+delta[i]**6/720

    U=np.zeros([N,N],dtype=complex)
    for i in range(N-1):
        U[i,i+1]= -4*Ec/d**2 +1j*4*Ec*ng/d

    L=np.zeros([N,N],dtype=complex)
    for i in range(N-1):
        L[i+1,i]= -4*Ec/d**2 -1j*4*Ec*ng/d
    H=U+D+L
    H[N-1,0]=-4*Ec/d**2 +1j*4*Ec*ng/d
    H[0,N-1]=-4*Ec/d**2 -1j*4*Ec*ng/d
    QH = qq.Qobj(H)
    # a,b=np.linalg.eig(H)
    # n=np.argsort(np.absolute(a))
    # n=np.argsort(a)
    # asort=a[n]
    asort = QH.eigenenergies()

    return asort[:40]
```

```python
sweep = np.linspace(0,1,30)
res = [test2(i) for i in sweep]
res = np.array(res)
plt.plot(sweep,res[:,:3])
```
